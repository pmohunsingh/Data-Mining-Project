```python
import numpy as np
import matplotlib.pyplot as plt

N = 3

#mode imputed ensemble, mode imputed bagged ensemble, train and test = train dataset 

# Mode
Accuracy_score = (82.71,83.34,88.18)
Precision_score = (53.53,54.89,68.30)
Recall_score = (66.69,68.33,81.23)
F1_score = (59.39, 60.88,74.21)


ind = np.arange(N, dtype='float')  # the x locations for the groups
width = 0.15      # the width of the bars

fig, ax = plt.subplots(figsize=(12,12))
rects1 = ax.bar(ind, Accuracy_score, width, color='#1926C4')
rects2 = ax.bar(ind + width, Precision_score, width, color='#C51979')
rects3 = ax.bar(ind + 2*width, Recall_score, width, color='#19C440')
rects4 = ax.bar(ind + 3*width, F1_score, width, color='#F3B119')

#text for labels, title and axes ticks
ax.set_ylabel('Score (In %)')
ax.set_title('Model Performance by Mode Imputation')
ax.set_xticks(ind + width*1.5)
ax.set_xticklabels(('Mode Imputed Ensemble', 'Mode Imputed Bagged Ensemble', 'Train Data = Test + Train'))

ax.legend((rects1[0], rects2[0], rects3[0], rects4[0]), ('Accuracy', 'Precision', 'Recall', 'F1-Score'))


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.00*height,
                '%.2f' % float(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

plt.show()
```
